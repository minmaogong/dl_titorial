```python
import torch
from torch.utils.data import Dataset, TensorDataset, DataLoader
```

```python
# 自定义数据集类
class MyDataset(Dataset):
    # 初始化
    def __init__(self, data):
        self.data = data

    # 返回数据集的大小
    def __len__(self):
        return len(self.data)

    # 返回指定索引的数据
    def __getitem__(self, idx):
        return self.data[idx]
```

```python
# 利用数据列表创建数据集
data = [1, 2, 3, 4, 5]
dataset = MyDataset(data)
```

```python
print(dataset[1])
```

## TensorDataset
