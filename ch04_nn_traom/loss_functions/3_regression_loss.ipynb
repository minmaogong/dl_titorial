```python
import torch
import torch.nn as nn
```

```python
# 定义预测输出值和目标值
output = torch.randn(5)
target = torch.randn(5)
print("Output:", output)
print("Target:", target)
```

## MAE (平均绝对误差 Mean Absolute Error，也叫 L1 Loss)

```python
loss_l1 = nn.L1Loss() # 对异常值鲁棒，但在0点处不可导
# 计算损失
loss = loss_l1(output, target)
print("Loss:", loss)
```

## MSE (均方误差 Mean Squared Error，也叫 L2 Loss)

```python
loss_mse = nn.MSELoss() # 对异常值敏感，遇到异常值时易发生梯度爆炸
# 计算损失
loss = loss_mse(output, target)
print("Loss:", loss)
```

## Smooth L1 Loss (平滑L1损失)

```python
loss_smooth_l1 = nn.SmoothL1Loss() # 当误差小时，使用L2 Loss，使得损失函数平滑可导。当误差较大时，使用L1 Loss降低异常值的影响
# 计算损失
loss = loss_smooth_l1(output, target)
print("Loss:", loss)
```
