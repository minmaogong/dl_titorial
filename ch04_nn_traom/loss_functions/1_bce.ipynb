```python
import torch
import torch.nn as nn
```

## 二元交叉熵损失函数（BCE）-- 用于二分类任务

```python
# 定义BCE损失函数（二元交叉熵损失函数）
loss_fn = nn.BCELoss()
```

### 输入一维数据（向量）

```python
# 定义5条数据
output = torch.randn(5)
print(output)
```

```python
# 计算预测概率--使用sigmoid函数将输出映射到0到1之间
proba = torch.sigmoid(output)
print("二分类预测概率: ", proba)
```

```python
# 定义标签
target = torch.tensor([0, 0, 1, 0, 1]).float()
```

```python
# 计算BCE损失
loss = loss_fn(proba, target)
print(loss)
```

### 输入二维数据

```python
# 定义数据
output = torch.randn(2, 3)
print(output)
```

```python
# 计算预测概率
proba = torch.sigmoid(output)
print("二分类预测概率: ", proba)
```

```python
# 定义标签
target = torch.tensor([[0, 0, 1], [0, 1, 1]]).float()
```

```python
# 计算BCE损失
loss = loss_fn(proba, target)
print(loss)
```

### 直接通过输出值计算损失

```python
loss_fn = nn.BCEWithLogitsLoss() # 内部包含了sigmoid 激活函数

# 计算损失
loss = loss_fn(output, target)
print(loss)
```
