```python
import torch
import torch.nn as nn
```

## 交叉熵损失函数 -- 用于多分类任务

```python
# 定义损失函数
loss_fn = nn.CrossEntropyLoss() # 内部包含了Softmax激活函数
```

```python
# 定义预测输出数据
output = torch.randn(8, 6) # 8条数据，6个分类
```

### 1. 目标值为标签编码（顺序编码）

```python
# 定义目标值
target = torch.randint(0, 6, (8,)) # 8条数据，每条数据的目标值为0-5之间的整数
print(target)
```

```python
# 计算交叉熵损失
loss = loss_fn(output, target)
print(loss)
```

### 2. 目标值为独热编码（One-hot编码）

```python
target_onehot = torch.zeros(8, 6) # 定义一个8x6的全零张量
```

```python
# 通过列表索引的方式将目标值对应位置的元素置为1，实现独热编码
target_onehot[ torch.arange(8), target ] = 1
```

```python
print(target_onehot)
```

```python
# 计算交叉熵损失
loss = loss_fn(output, target_onehot)
print(loss)
```
