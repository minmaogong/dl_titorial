```python
import torch
import torch.nn as nn
```

```python
# 定义数据
x = torch.tensor(10.0)
y = torch.tensor([3.0])
```

```python
# 定义参数（权重和偏置）
w =  torch.randn(1, requires_grad=True) # 权重
b = torch.randn(1, requires_grad=True) # 偏置
```

```python
# 前向传播，计算输出值
z = w * x + b
print("输出值: ", z)
```

```python
# 定义损失函数
loss_fn = nn.MSELoss()
# 计算损失
loss = loss_fn(z, y)
print("损失值: ", loss)
```

```python
# 反向传播
loss.backward()
```

```python
# 查看参数梯度
print("权重梯度: ", w.grad)
print("偏置梯度: ", b.grad)
```

```python
print(x.is_leaf)
print(y.is_leaf)
print(w.is_leaf)
print(b.is_leaf)
print(z.is_leaf) # 通过计算得到的张量不是叶子节点
print(loss.is_leaf) # 通过计算得到的张量不是叶子节点
```
