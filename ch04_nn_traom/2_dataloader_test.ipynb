```python
import torch
from torch.utils.data import TensorDataset, DataLoader
```

```python
# 定义数据
X = torch.randn(10, 3) # 10个样本，每个样本3个特征
y = torch.randn(10) # 10个样本的标签
print(X)
print(y)
```

```python
# 定义数据集
dataset = TensorDataset(X, y)
```

```python
print(dataset[2]) # 从X和y中取出索引为2的样本
```

```python
# 创建数据加载器
# batch_size: 每个批次的样本数
# shuffle: 是否打乱数据
# drop_last: 如果数据集大小不能被batch_size整除，是否丢弃
dataloader = DataLoader(dataset=dataset, batch_size=4, shuffle=True, drop_last=True)
```

```python
# 遍历数据
for x_batch, y_batch in dataloader:
    print(x_batch)
    print(y_batch)
    print('---')
```
