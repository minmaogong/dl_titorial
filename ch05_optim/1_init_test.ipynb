```python
import torch
import torch.nn as nn
```

```python
# 定义一个全连接层
linear = nn.Linear(3, 5)
```

## 常数初始化

```python
nn.init.zeros_(linear.weight)
nn.init.zeros_(linear.bias)
```

```python
print(linear.weight)
print(linear.bias)
```

```python
nn.init.ones_(linear.weight)
print(linear.weight)
```

```python
nn.init.constant_(linear.bias, 10)
print(linear.bias)
```

## 秩初始化

```python
nn.init.eye_(linear.weight) # 用单位矩阵初始化权重
print(linear.weight)
```
