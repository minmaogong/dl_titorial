```python
import torch
import torch.nn as nn
```

```python
# 定义一个全连接层
linear = nn.Linear(3, 5)
```

## 常数初始化

```python
nn.init.zeros_(linear.weight)
nn.init.zeros_(linear.bias)
```

```python
print(linear.weight)
print(linear.bias)
```

```python
nn.init.ones_(linear.weight)
print(linear.weight)
```

```python
nn.init.constant_(linear.bias, 10)
print(linear.bias)
```

## 秩初始化

```python
nn.init.eye_(linear.weight) # 用单位矩阵初始化权重
print(linear.weight)
```

## 随机初始化

```python
# 正态分布初始化
nn.init.normal_(linear.weight, mean=0, std=1) # 用均值为0，标准差为1的正态分布初始化权重
print(linear.weight)
```

```python
nn.init.normal_(linear.weight, mean=5, std=2)
print(linear.weight)
```

```python
# 均匀分布初始化
nn.init.uniform_(linear.weight, a=0, b=10) # 用均匀分布初始化权重，范围为[a, b)
print(linear.weight)
```

## Xavier初始化

```python
# Xavier初始化参数适用于Sigmoid和Tanh等激活函数，能有效缓解梯度消失或爆炸问题
```

```python
nn.init.xavier_normal_(linear.weight) # 用Xavier正态分布初始化权重
print(linear.weight)
```

```python
nn.init.xavier_uniform_(linear.weight) # 用Xavier均匀分布初始化权重
print(linear.weight)
```
