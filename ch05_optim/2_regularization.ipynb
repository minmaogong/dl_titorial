```python
import torch
import torch.nn as nn
from torch import optim
```

# 正则化

```python
x = torch.randint(0, 10, (5, 3)).float()
print(x)
```

## 1.Batch Normalization 批量标准化 调整各层的激活值分布使其拥有适当的广度，BN层通常放在线性层（全连接层/卷积层）之后，激活函数之前。

```python
# 定义 BN 层 （Batch Normalization 批量标准化）
bn = nn.BatchNorm1d(3)
```

```python
y = bn(x)
print(y)
```

## 2. 权值衰减 通过在学习的过程中对大的权重进行“惩罚”，可以有效的抑制过拟合

```python
model = nn.Linear(1, 3)
optimizer = optim.SGD(model.parameters(), lr=0.1, weight_decay=1e-2) # weight_decay: 正则化系数
```

## 3. Dropout 随机失活（暂退法） 在训练过程中随机的将一部分神经元的输出置为0（关闭神经元），防止神经元之间过度依赖，通常放在激活函数之后，下一个线性层（全连接层/卷积层）之前

```python
x = torch.randint(1, 10, (10, )).float()
print(x)
```

```python
# 定义 Dropout 层
dropout = nn.Dropout(p=0.5) # p: 失活概率 每个神经元以0.5的概率关闭
```

```python
y = dropout(x)
print(y) # 没关闭的神经元，会进行缩放，按 1/(1-p) = 1(1-0.5) = 2 的比例放大
```
