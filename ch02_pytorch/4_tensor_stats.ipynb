```python
import torch
```

```python
tensor1 = torch.randint(1, 10, (2, 3, 4)) # 2个3行4列的矩阵
print(tensor1)
```

```python
# 1. sum 求和
print(tensor1.sum()) # 求所有元素的和
```

```python
print(tensor1.sum(dim=0)) # dim=0: 表示2个矩阵对应元素相加，得到一个3行4列的矩阵 dim=1: 表示每个矩阵的每列元素相加，每个矩阵变成1维向量，最终得到一个2行4列的矩阵 dim=2: 表示每个矩阵的每行元素相加，每个矩阵变成1维向量，最终得到一个2行3列的矩阵
```

```python
print(torch.sum(tensor1, dim=1)) # 等价于 tensor1.sum(dim=1)
```

```python
# 2. mean 求平均值
tensor1 = tensor1.float()
print(tensor1.mean()) # 求所有元素的平均值
print(tensor1.mean(dim=0)) # dim=0: 表示2个矩阵对应元素求平均值，得到一个3行4列的矩阵
print(tensor1.mean(dim=1)) # dim=1: 表示每个矩阵的每列元素求平均值，每个矩阵变成1维向量，最终得到一个2行4列的矩阵
print(tensor1.mean(dim=2)) # dim=2: 表示每个矩阵的每行元素求平均值，每个矩阵变成1维向量，最终得到一个2行3列的矩阵
```

```python
# 3. max / min 求最大值/最小值
print(tensor1.max()) # 求所有元素的最大值
print(tensor1.max(dim=0)) # dim=0: 表示2个矩阵对应元素求最大值，得到一个3行4列的矩阵
print(tensor1.max(dim=1)) # dim=1: 表示每个矩阵的每列元素求最大值，每个矩阵变成1维向量，最终得到一个2行4列的矩阵
print(tensor1.max(dim=2)) # dim=2: 表示每个矩阵的每行元素求最大值，每个矩阵变成1维向量，最终得到一个2行3列的矩阵
```

```python
print(tensor1.argmax(dim=0)) # dim=0: 表示2个矩阵对应元素求最大值的索引
print(tensor1.argmax(dim=1)) # dim=1: 表示每个矩阵的每列元素求最大值的索引
print(tensor1.argmax(dim=2)) # dim=2: 表示每个矩阵的每行元素求最大值的索引
```

```python
# 4. var, std 求方差/标准差
print(tensor1.std(dim=1)) # 标准差
```

```python
print(tensor1.var(dim=1)) # 方差
```

```python
# 5. unique 去重
print(tensor1.unique())
```

```python
# 6. sort 排序
print(tensor1.sort()) # 默认dim=-1, 表示对最后一个维度每一个元素进行排序，等于对每一行内部排序
print(tensor1.sort(dim=1)) # dim=1, 表示对每个矩阵的每列元素进行排序
```
