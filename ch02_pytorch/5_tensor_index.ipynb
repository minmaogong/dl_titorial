```python
import torch
```

```python
tensor1 = torch.randint(1, 10, (3, 5, 4)) # 创建一个3*5*4的张量，元素为1-9之间的随机整数
print(tensor1)
```

## 简单索引

```python
print(tensor1[0, 1, 2]) # 取0维第0个，1维第1个，2维第2个对应的元素
```

```python
print(tensor1[0, 0]) # 取0维第0个，1维第0个，2维所有元素
print(tensor1[0, 0, :]) # 取0维第0个，1维第0个，2维所有元素
```

```python
print(tensor1[1]) # 取0维第1个所有元素
```

```python
print(tensor1[1, :, 2]) # 取0维第1个，1维所有，2维第2个对应的所有元素
```

```python
print(tensor1[:, 1]) # 取0维第0个、第1个、第2个，1维第1个所有的元素
```

```python
print(tensor1[:, :, 1]) # 取0维第0个-第2个，1维第0-第4个，2维第1个所有元素
```

## 范围索引（切片索引）

```python
print(tensor1[1:]) # 取0维第1个后面的所有元素 与 tensor1[1, :] 不同，tensor1[1, :] 表示取0维第1个所有的元素
```

```python
print(tensor1[:, 1:4, 0:3]) # 取0维第0个-第2个，1维第[1, 4)个，2维第[0, 3) 对应的所有元素
```

```python
print(tensor1[:, 1:4, 0:3:2]) #  0:3:2 表示2维取 start=0 end=3 step=2 索引对应的元素
```

```python
print(tensor1[-2:-1, 1:4, 0:3:2]) # -2:-1 表示0维取index=-2（-2:-1 切片是不包含end的）index=-2 == index=1
```

## 列表索引

```python
list1 = [0, 2] # 0维 第0个 和 第2个， 即0维对应的坐标
list2 = [1, 4] # 0维第0个1维第1个 0维第2个1维第4个， 即0维下1维对应的坐标 等价于：tensor1[0, 1, :] 和 tensor1[2, 4, :]
print(tensor1[list1, list2])
```

```python
list1 = [[0], [1]] #取0维度索引维0和1
list2 = [1, 4, 3]
print(tensor1[list1, list2])
```
