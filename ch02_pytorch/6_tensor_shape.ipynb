```python
import torch
```

```python
tensor1 = torch.randint(1, 10, (3, 2, 6))
print(tensor1)
```

## 变换维度

```python
# 1. .T 转置所有维度 0维和2维的交换
print(tensor1.T) # (3, 2, 6) -> (6, 2, 3)
```

```python
# 2. .mT 矩阵转换
print(tensor1.mT)
```

```python
# 3. transpose 交换两个维度
print(tensor1.transpose(0, 2)) # (3, 2, 6) -> (6, 2, 3) 0维和2维交换
```

```python
print(tensor1.transpose(-1, -2)) # (3, 2, 6) -> (3, 6, 2) 1维和2维交换
```

```python
# 4. permute 重新排列所有维度
print(tensor1.permute(2, 0, 1)) # (3, 2, 6) -> (6, 3, 2) 0维变成1维，1维变成2维，2维变成0维
```

## 调整形状

```python
# 1. reshape 调整形状
print(tensor1.shape)
print(tensor1.reshape(4, 3, 3)) # (3, 2, 6) -> (4, 3, 3) 重新调整形状，元素总数不变
```

```python
print(tensor1.reshape(4, 9)) # (3, 2, 6) -> (4, 9) 重新调整形状，元素总数不变
```

```python
print(tensor1.reshape(6, -1)) # (3, 2, 6) -> (6, 6) 重新调整形状，元素总数不变 -1表示自动计算该维度的大小
```

```python
print(tensor1.reshape(-1)) # (3, 2, 6) -> (36,) 重新调整形状，元素总数不变 变换成一个行向量
```

```python
print(tensor1.reshape(1, -1)) # (3, 2, 6) -> (1, 36) 重新调整形状，元素总数不变 变换成一个1x36的二维矩阵
```

```python
print(tensor1.reshape(-1, 1)) # (3, 2, 6) -> (36, 1) 重新调整形状，元素总数不变 变换成一个36x1的二维矩阵
```

```python
print(tensor1.reshape(2, 3, 3, 2)) # (3, 2, 6) -> (2, 3, 3, 2) 重新调整形状，元素总数不变
```

```python
# 2. view 调整形状 需要内存连续。共享内存
print(tensor1.shape)
```

```python
print(tensor1.view(4, -1))
```

```python
print(tensor1.is_contiguous()) # 判断是否内存连续
```

```python
tensor1 = tensor1.permute(2, 0 ,1) # 维度变换，会导致内存不连续
print(tensor1.shape)
```

```python
print(tensor1.is_contiguous()) # 判断是否内存连续
```

```python
print(tensor1.view(4, -1)) # 会报错，因为内存不连续
```

```python
print(tensor1.contiguous().view(4, -1)) # 先调用 contiguous() 方法，使内存连续，再调用 view() 方法调整形状
```

## 增删维度

```python
tensor1 = torch.randint(1, 10, (3, 2, 6))
print(tensor1)
```

```python
# 1. 增加维度
print(tensor1.reshape(1, 3, 2, 6)) # (6, 3, 2) -> (1, 6, 3, 2) 增加一个维度
```

```python
print(tensor1.reshape(3, 1, 2, 6))
```

```python
print(tensor1.unsqueeze(0)) # 0维增加一个维度 (3, 2, 6) -> (1, 3, 2, 6)
```

```python
print(tensor1.unsqueeze(1)) # 1维增加一个维度 (3, 2, 6) -> (3, 1, 2, 6)
#
```

```python
print(tensor1.unsqueeze(dim=3)) # 3维增加一个维度 (3, 2, 6) -> (3, 2, 6, 1)
```

```python
tensor2 = tensor1.unsqueeze(0) # 0维增加一个维度 (3, 2, 6) -> (1, 3, 2, 6)
print(tensor2)
```

```python
# 2. 删除维度(降维)
print(tensor2.squeeze(0).shape) # 删除0维 (1, 3, 2, 6) -> (3, 2, 6)
print(tensor2.squeeze().shape) # 删除所有为1的维度 (1, 3, 2, 6) -> (3, 2, 6)
```

## 张量拼接

```python
# 1. cat 指定维度拼接 要求只有1个维度不同，其他维度相同
tensor1 = torch.randint(1, 10, (3, 2, 6))
tensor2 = torch.randint(1, 10, (3, 3, 6))
print(tensor1)
print(tensor2)
```

```python
tensor3 = torch.cat([tensor1, tensor2], dim=1) # (3, 2, 6) + (3, 3, 6) -> (3, 5, 6) 指定维度拼接
print(tensor3)
print(tensor3.shape)
```

```python
# 2. stack 堆叠拼接 要求所有维度相同（形状相同）
tensor1 = torch.randint(1, 10, (4, 3, 6))
tensor2 = torch.randint(1, 10, (4, 3, 6))
print(tensor1)
print(tensor2)
```

```python
print(torch.stack([tensor1, tensor2]).shape) # 默认dim=0 以0维为单位进行堆叠拼接 (4, 3, 6) + (4, 3, 6) -> (2, 4, 3, 6)
print(torch.stack([tensor1, tensor2], dim=1).shape) # dim=1 以1维为单位进行堆叠拼接 (4, 3, 6) + (4, 3, 6) -> (4, 2, 3, 6)
print(torch.stack([tensor1, tensor2], dim=2).shape) # dim=2 以2维为单位进行堆叠拼接 (4, 3, 6) + (4, 3, 6) -> (4, 3, 2, 6)
print(torch.stack([tensor1, tensor2], dim=3).shape) # dim=3 张量的每一个元素进行堆叠拼接 (4, 3, 6) + (4, 3, 6) -> (4, 3, 6, 2)
```
