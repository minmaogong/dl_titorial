```python
import torch
```

```python
tensor1 = torch.randint(1, 10, (3, 2, 6))
print(tensor1)
```

## 变换维度

```python
# 1. .T 转置所有维度 0维和2维的交换
print(tensor1.T) # (3, 2, 6) -> (6, 2, 3)
```

```python
# 2. .mT 矩阵转换
print(tensor1.mT)
```

```python
# 3. transpose 交换两个维度
print(tensor1.transpose(0, 2)) # (3, 2, 6) -> (6, 2, 3) 0维和2维交换
```

```python
print(tensor1.transpose(-1, -2)) # (3, 2, 6) -> (3, 6, 2) 1维和2维交换
```

```python
# 4. permute 重新排列所有维度
print(tensor1.permute(2, 0, 1)) # (3, 2, 6) -> (6, 3, 2) 0维变成1维，1维变成2维，2维变成0维
```

## 调整形状

```python
# 1. reshape 调整形状
print(tensor1.shape)
print(tensor1.reshape(4, 3, 3)) # (3, 2, 6) -> (4, 3, 3) 重新调整形状，元素总数不变
```

```python
print(tensor1.reshape(4, 9)) # (3, 2, 6) -> (4, 9) 重新调整形状，元素总数不变
```

```python
print(tensor1.reshape(6, -1)) # (3, 2, 6) -> (6, 6) 重新调整形状，元素总数不变 -1表示自动计算该维度的大小
```

```python
print(tensor1.reshape(-1)) # (3, 2, 6) -> (36,) 重新调整形状，元素总数不变 变换成一个行向量
```

```python
print(tensor1.reshape(1, -1)) # (3, 2, 6) -> (1, 36) 重新调整形状，元素总数不变 变换成一个1x36的二维矩阵
```

```python
print(tensor1.reshape(-1, 1)) # (3, 2, 6) -> (36, 1) 重新调整形状，元素总数不变 变换成一个36x1的二维矩阵
```

```python
# 2. view 调整形状 需要内存连续。共享内存
print(tensor1.shape)
```

```python
print(tensor1.view(4, -1))
```

```python
print(tensor1.is_contiguous()) # 判断是否内存连续
```

```python
tensor1 = tensor1.permute(2, 0 ,1) # 维度变换，会导致内存不连续
print(tensor1.shape)
```

```python
print(tensor1.is_contiguous()) # 判断是否内存连续
```

```python
print(tensor1.view(4, -1)) # 会报错，因为内存不连续
```

```python
print(tensor1.contiguous().view(4, -1)) # 先调用 contiguous() 方法，使内存连续，再调用 view() 方法调整形状
```
