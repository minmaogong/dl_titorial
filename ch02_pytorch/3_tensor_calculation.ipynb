```python
import torch
```

```python
tensor1 = torch.randint(low=1, high=10, size=(2, 3))
print(tensor1)
```

## 1. 张量基本运算

```python
# 使用运算符号 (不改变原张量)
print(tensor1 + 10) # tensor1 的每个元素都加上10
print(tensor1) # tensor1 不会发生变化
```

```python
# 调用 add 方法 （不改变原张量）
print(tensor1.add(20)) # tensor1 的每个元素都加上20
print(tensor1) # tensor1 不会发生变化
```

```python
# 使用 add_ 方法 （改变原张量）
print(tensor1.add_(30)) # tensor1 的每个元素都加上30
print(tensor1) # tensor1 会发生变化
```

```python
tensor2 = torch.tensor([1, 2, 3])
print(tensor2.exp()) # 已 e 为底数求幂(指数) e^x
```

```python
print(tensor2.log()) # 已 e 为底数求对数 log_e(x)
```

## 2. 元素级运算（张量与张量的运算）

```python
tensor1 = torch.randint(low=1, high=10, size=(2, 3))
tensor2 = torch.randint(low=1, high=10, size=(2, 3))
print(tensor1)
print(tensor2)
```

```python
print(tensor1 + tensor2) # 张量对应元素相加
```

```python
tensor3 = torch.randint(low=1, high=10, size=(1, 3))
print(tensor3)
```

```python
print(tensor1 + tensor3) # 张量对应元素相加，广播机制
```

```python
print(tensor1 * tensor2) # 张量对应元素相乘-哈达玛积（元素级乘法）
```

## 3. 张量的矩阵乘法

```python
tensor1 = torch.randint(low=1, high=10, size=(2, 3))
tensor2 = torch.randint(low=1, high=10, size=(2, 3))
print(tensor1)
print(tensor2.T)
```

```python
print(tensor1 @ tensor2.T) #
```

```python
print(tensor1.mm(tensor2.T)) # mm: matrix multiplication 矩阵乘法，只能用于二维张量
```

```python
print(tensor1.matmul(tensor2.T)) # matmul: matrix multiplication 矩阵乘法，可以用于二维张量和高维张量
```

```python
tensor1 = torch.randint(1, 10, (2, 3, 4)) # 2个3行4列的矩阵
tensor2 = torch.randint(1, 10, (2, 4, 2))
print(tensor1)
print(tensor2)
```

```python
print(tensor1 @ tensor2) # 得到一个(2, 3, 2) 的张量
```

```python
print(tensor1.matmul(tensor2)) # 得到一个(2, 3, 2) 的张量
```
