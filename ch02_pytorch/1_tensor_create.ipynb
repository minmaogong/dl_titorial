# 张量创建

```python
import numpy as np
import torch
```

## 1. 基于内容创建张量

```python
# 传入一个标量
tensor1 = torch.tensor(10)
print(tensor1)
```

```python
print(tensor1.shape)
```

```python
# 传入一个列表
tensor2 = torch.tensor([1, 2, 3, 4, 5])
print(tensor2)
print(tensor2.shape)
```

```python
# 传入二维列表（矩阵）
tensor3 = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(tensor3)
print(tensor3.shape)
```

```python
# 传入高维列表
tensor4 = torch.tensor([[[1, 2, 3], [4, 5, 6]], [[70, 80, 90], [100, 110, 120]]])
print(tensor4)
print(tensor4.shape) # torch.Size([2, 2, 3]) 表示张量由2个矩阵，每个矩阵有2行3列组成
```

```python
print(tensor4.dtype)
```

```python
# 传入数据类型为浮点型
tensor5 = torch.tensor([1, 2, 3.0])
print(tensor5)
print(tensor5.dtype)
```

```python
# 传入一个ndarray
tensor6 = torch.tensor(np.array([1, 2, 3.0]))
print(tensor6)
print(tensor6.dtype)
```

## 2. 基于形状创建张量

```python
tensor1 = torch.Tensor(3, 2, 4) # 创建一个3维张量，形状为(3, 2, 4)，即3个矩阵，每个矩阵有2行4列组成
print(tensor1)
print(tensor1.shape)
```

```python
tensor2 = torch.Tensor(10) # 创建一个1维张量，形状为(10,)，即10个元素组成
print(tensor2)
print(tensor2.shape)
```

```python
print(tensor2.dtype)
```

```python
# 也可以指定内容创建张量
tensor3 = torch.Tensor([[1, 2, 3], [4, 5, 6]])
print(tensor3)
print(tensor3.shape)
```

```python
print(tensor3.dtype)
```
