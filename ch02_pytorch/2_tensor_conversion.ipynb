```python
import torch
import numpy as np
```

# 张量转换

## 元素类型转换

```python
tensor1 = torch.tensor([1, 2, 3])
print(tensor1.dtype)
```

```python
# type 方法
tensor1 = tensor1.type(torch.float32)
print(tensor1.dtype)
```

```python
# 直接调用类型方法
tensor1 = tensor1.double() # 转换为 double 类型
print(tensor1.dtype)
```

```python
# to方法
tensor1 = tensor1.to(torch.int64)
print(tensor1.dtype)
```

## Tensor 与 ndarray 转换

```python
np.set_printoptions(precision=6) # 设置打印精度为6位小数
torch.set_printoptions(precision=6) # 设置打印精度为6位小数
```

```python
# 1. Tensor 转 ndarray
tensor1 = torch.rand(3, 2) # 创建一个3行2列的随机张量
print(tensor1)
```

```python
ndarray1 = tensor1.numpy() # 将张量转换为ndarray
print(ndarray1)
```

```python
# 验证数据内存共享
tensor1[:, 0] = 5 # 将张量的第一列全部赋值为5 ":" 表示所有行， "1:" 表示从第二行开始的所有行 "0:2" 表示第一行和第二行，可以简化成":2"
print(tensor1)
print(ndarray1) # ndarray 也会发生变化
ndarray1[:2, 1] = 10 # 将ndarray的前两行的第二列全部赋值为10
print(tensor1)
print(ndarray1) ## 张量也会发生变化
```

```python
# 避免内存共享
ndarray2 = tensor1.numpy().copy() # 将张量转换为ndarray，并复制一份数据，避免内存共享
tensor1[:, 1] = 100 # 将张量的第二列全部赋值为100
print(tensor1)
print(ndarray2) # ndarray 不会发生变化
```

```python
# 2. ndarray 转 Tensor
ndarray1 = np.random.randn(3, 2) # 创建一个3行2列的随机ndarray
print(ndarray1)
```

```python
tensor1 = torch.from_numpy(ndarray1) # 将ndarray转换为张量
print(tensor1)
```

```python
tensor1[2, 1] = 5 # 将张量的第三行第二列赋值为5
print(tensor1)
print(ndarray1) # ndarray 也会发生变化
```

```python
# 避免内存共享
tensor2 = torch.from_numpy(ndarray1).clone() # 将ndarray转换为张量，并复制一份数据，避免内存共享
print(tensor2)
```

```python
tensor1[2, 0] = 1.25
print(tensor1)
print(ndarray1) # ndarry 也会发生变化
print(tensor2) # 张量不会发生变化
```

```python
tensor2[2, 1] = 5.2
print(tensor2)
print(ndarray1)
```

```python
# 3. 用tensor 方法直接基于内容创建张量
ndarray1 = np.random.randn(3, 2)
print(ndarray1)
```

```python
tensor1 = torch.tensor(ndarray1) # 将ndarray转换为张量，不会共享内存
print(tensor1)
```

```python
tensor1[2, 0] = 1.25
print(tensor1)
print(ndarray1) # ndarray 不会发生变化
```
