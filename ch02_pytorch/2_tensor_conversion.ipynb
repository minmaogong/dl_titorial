# 张量转换

## 元素类型转换

```python
tensor1 = torch.tensor([1, 2, 3])
print(tensor1.dtype)
```

```python
# type 方法
tensor1 = tensor1.type(torch.float32)
print(tensor1.dtype)
```

```python
# 直接调用类型方法
tensor1 = tensor1.double() # 转换为 double 类型
print(tensor1.dtype)
```

```python
# to方法
tensor1 = tensor1.to(torch.int64)
print(tensor1.dtype)
```
