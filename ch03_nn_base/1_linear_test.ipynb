```python
import torch
import torch.nn as nn
```

```python
# 定义一个线性层，输入特征数为2，输出特征数为3 (定义了一个有2个输入神经元、3个输出神经元的全连接层)
linear = nn.Linear(in_features=2, out_features=3)
```

```python
print("权重：", linear.weight) # [[w11, w21], [w12, w22], [w13, w23]].T
```

```python
print("偏置：", linear.bias) # b1 b2 b3
```

## 前向传播

```python
import torch

# 输入数据
x = torch.Tensor([1, 2]) # 输入特征数为2
```

```python
# 前向传播，得到输出
y = linear.forward(x)
print(y) # y1 = w11*x1 + w21*x2 + b1, y2 = w12*x1 + w22*x2 + b2, y3 = w13*x1 + w23*x2 + b3
```

```python
# 输入多条数据
X = torch.randn(5, 2) # 5条数据，每条数据有2个特征
```

```python
y = linear.forward(X) # 输入5条数据，得到5条输出，每条输出有3个特征 5x3
print(y)
```

```python
# 输入高维数据
X = torch.randn(10, 5, 2) # 10个样本，每个样本有5条数据，每条数据有2个特征 10x5x2
```

```python
y = linear.forward(X) # 输入10个样本，每个样本有5条数据，每条数据有2个特征，得到10个样本，每个样本有 10x5x3
print(y)
```

```python
y = linear(X) # 与调用forward方法等价
print(y.shape)
```
