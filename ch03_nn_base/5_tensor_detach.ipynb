## detach 对张量的影响

```python
# 定义数据
x = torch.tensor(2.0, requires_grad=True)
print(x)
```

```python
y = x.detach()
print(y)
```

```python
print(x.requires_grad)
print(y.requires_grad)
```

```python
print(x.untyped_storage().data_ptr())
print(y.untyped_storage().data_ptr())
```

```python
# 分别基于新，y进行计算
z1 = x ** 2
z2 = y ** 2
```

```python
print(z1)
print(z2)
```

```python
z1.backward()
```

```python
z2.backward() # 会报错，因为y不需要梯度
```

## detach后参与计算的影响

```python
# 定义输入数据
x = torch.ones(2, 2, requires_grad=True)
```

```python
# 定义后续运算
y = x ** 2
print(y)
```

```python
# 分离
u = y.detach() + torch.ones_like(y)
print(u)
```

```python
# 让u与x进行计算，得到输出
z = u * x
print(z)
```

```python
# 反向传播
z.sum().backward()
```
